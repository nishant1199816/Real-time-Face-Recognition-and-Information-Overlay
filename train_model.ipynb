```python
import face_recognition
import os
import pickle

IMAGE_DIR = "Images" 
known_encodings = []
known_names = []

print("Scanning folders... Ismein thoda time lag sakta hai.")
# Har folder se images read karna

for person_name in os.listdir(IMAGE_DIR):
    folder = os.path.join(IMAGE_DIR, person_name)
    if os.path.isdir(folder):
        for image_name in os.listdir(folder):
            if image_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_path = os.path.join(folder, image_name)
                image = face_recognition.load_image_file(image_path)
                encodings = face_recognition.face_encodings(image)
                if len(encodings) > 0:
                    known_encodings.append(encodings[0])
                    known_names.append(person_name)
                    print(f"Success: {person_name} extracted")

# Pickle file save karna

data = {"encodings": known_encodings, "names": known_names}

with open("encodings.pickle", "wb") as f:

    f.write(pickle.dumps(data))

print("\nSuccess! 'encodings.pickle' file ban gayi hai.")
```
